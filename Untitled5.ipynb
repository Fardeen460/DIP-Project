```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image_path = list(uploaded.keys())[0]  # Get the uploaded file name
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for display
```

```python
plt.figure(figsize=(6, 6))
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
def adjust_brightness(image, value):
    """
    Adjusts the brightness of the image.
    :param image: Input image.
    :param value: Brightness adjustment value (positive or negative).
    :return: Brightness-adjusted image.
    """
    adjusted_image = cv2.convertScaleAbs(image, alpha=1, beta=value)
    return adjusted_image

# Example: Increase brightness by 50
brightness_adjusted = adjust_brightness(image, 50)

# Display the result
plt.figure(figsize=(6, 6))
plt.title("Brightness Adjusted Image")
plt.imshow(brightness_adjusted)
plt.axis('off')
plt.show()
```

```python
def adjust_contrast(image, alpha):
    """
    Adjusts the contrast of the image.
    :param image: Input image.
    :param alpha: Contrast adjustment factor (e.g., 1.5 for higher contrast, 0.5 for lower contrast).
    :return: Contrast-adjusted image.
    """
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=0)
    return adjusted_image

# Example: Increase contrast by 1.5
contrast_adjusted = adjust_contrast(image, 1.5)

# Display the result
plt.figure(figsize=(6, 6))
plt.title("Contrast Adjusted Image")
plt.imshow(contrast_adjusted)
plt.axis('off')
plt.show()
```

```python
def remove_noise(image, method='gaussian', kernel_size=5):
    """
    Removes noise from the image.
    :param image: Input image.
    :param method: Noise removal method ('gaussian' or 'median').
    :param kernel_size: Size of the kernel for blurring.
    :return: Denoised image.
    """
    if method == 'gaussian':
        denoised_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
    elif method == 'median':
        denoised_image = cv2.medianBlur(image, kernel_size)
    else:
        raise ValueError("Invalid method. Use 'gaussian' or 'median'.")
    return denoised_image

# Example: Remove noise using Gaussian Blur
denoised_image = remove_noise(image, method='gaussian', kernel_size=5)

# Display the result
plt.figure(figsize=(6, 6))
plt.title("Denoised Image (Gaussian Blur)")
plt.imshow(denoised_image)
plt.axis('off')
plt.show()
```

```python
def simple_image_editor(image, brightness=0, contrast=1.0, denoise=False, denoise_method='gaussian', kernel_size=5):
    """
    A simple image editor that performs brightness, contrast adjustment, and noise removal.
    :param image: Input image.
    :param brightness: Brightness adjustment value.
    :param contrast: Contrast adjustment factor.
    :param denoise: Whether to apply noise removal.
    :param denoise_method: Noise removal method ('gaussian' or 'median').
    :param kernel_size: Size of the kernel for noise removal.
    :return: Edited image.
    """
    # Adjust brightness
    edited_image = adjust_brightness(image, brightness)

    # Adjust contrast
    edited_image = adjust_contrast(edited_image, contrast)

    # Remove noise (if enabled)
    if denoise:
        edited_image = remove_noise(edited_image, method=denoise_method, kernel_size=kernel_size)

    return edited_image

# Example: Apply brightness, contrast, and noise removal
edited_image = simple_image_editor(image, brightness=30, contrast=1.2, denoise=True, denoise_method='gaussian', kernel_size=5)

# Display the result
plt.figure(figsize=(6, 6))
plt.title("Edited Image")
plt.imshow(edited_image)
plt.axis('off')
plt.show()
```
